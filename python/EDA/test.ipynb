```python
import pandas as pd
file_path = "C:/Users/santh/Downloads/archive (3)/StudentsPerformance.csv"

data = pd.read_csv(file_path)

```

```python
print(data.isnull().sum())
```

```python
data.columns = data.columns.str.lower().str.replace(' ', '_')

```

```python
data['gender'] = data['gender'].astype('category')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
print(data.describe())
```

```python
plt.figure(figsize=(8,6))
sns.histplot(data['math_score'], bins=20, kde=True)
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()

```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='gender', y='writing_score', data=data)
plt.title('Gender vs Writing Scores')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='math_score', y='reading_score', data=data, hue='gender')
plt.title('Math Score vs Reading Score')
plt.show()
```
